Q1) What is the difference between a function and a method in Python?

Ans)In Python, both functions and methods are blocks of reusable code designed to perform specific tasks, but their key distinction lies in their relationship to classes and objects:

**Function:** A function is a standalone block of code that is defined independently of any class. It can be called directly by its name and operates on the data passed to it as arguments. Functions are generally used for independent tasks or reusable logic that doesn't inherently belong to a specific object or its data.
example:  

            def greet(name):
        return f"Hello, {name}!"

    message = greet("Alice")
    print(message)


**Method:** A method is a function that is defined within a class. It is associated with objects (instances) of that class and is designed to operate on the object's data (attributes). Methods are called on an object using dot notation (e.g., object.method()) and implicitly receive the object itself as their first argument (conventionally named self).

example:  class Dog:
        def __init__(self, name):
            self.name = name

        def bark(self):
            return f"{self.name} says Woof!"

    my_dog = Dog("Buddy")
    sound = my_dog.bark()
    print(sound)

Q4)What is the purpose of the `return` statement in a Python function?

Ans)The return statement in a Python function serves two primary purposes:

**Exiting the Function:**
When a return statement is encountered within a function, the function's execution immediately terminates at that point. Any code following the return statement within the function will not be executed.

**Sending a Value Back to the Caller:**
The return statement allows a function to send a value or set of values back to the part of the code that called it. This returned value can then be used for further processing, assigned to a variable, passed as an argument to another function, or otherwise utilized within the program. If no value is explicitly specified with the return statement, or if no return statement is present in the function, the function implicitly returns None.

In essence, the return statement enables functions to not only perform actions but also to communicate results or data back to the rest of the program, making them powerful and reusable building blocks for code organization and functionality.

Example:

    def add_numbers(a, b):
    return a + b

    Calling the function

Q5)What are iterators in Python and how do they differ from iterables?

Ans)In Python, iterables and iterators are fundamental concepts for handling sequences of data.

**An iterable** is an object that can be iterated over, meaning it can return its members one at a time. Common examples of iterables include lists, tuples, strings, sets, and dictionaries. An object is considered iterable if it implements either the __iter__() method, which returns an iterator, or the __getitem__() method, which allows access to elements by index. You can typically use an iterable directly in a for loop.


**An iterator** is an object that represents a stream of data and allows you to traverse through the elements of an iterable one by one. It implements the iterator protocol, which consists of two methods:
__iter__(): This method returns the iterator object itself.
__next__(): This method returns the next item from the sequence. When there are no more items, it raises a StopIteration exception.

Example:

    my_list = [10, 20, 30, 40]  # This list is an iterable
    for item in my_list:
    print(item)

Q6) Explain the concept of generators in Python and how they are defined?

Ans)**Generators** in Python are a special type of iterable that allow for the creation of iterators in a memory-efficient manner. They produce a sequence of values lazily, one at a time, rather than storing all values in memory simultaneously. This characteristic makes them particularly useful when dealing with large datasets or infinite sequences where loading the entire dataset into memory would be impractical or impossible.
How Generators are Defined:
Generators are defined in Python using functions, but instead of using the return keyword to return a value and terminate the function, they use the yield keyword.


def my_generator_function():
    # Some code
    yield value1
    # More code
    yield value2
    # Even more code
    yield value3

Q7) What are the advantages of using generators over regular functions?

Ans) Generators offer several advantages over regular functions, particularly when dealing with sequences of data:

**Memory Efficiency:**
Generators produce values one at a time using the yield keyword, rather than computing and storing an entire list or sequence in memory. This makes them highly memory-efficient, especially when working with large datasets or potentially infinite sequences, as only the current value is held in memory.

**Lazy Evaluation:**
Values are computed and yielded only when requested, on demand. This "lazy" evaluation improves performance by avoiding unnecessary computations when not all elements of a sequence are required.

**Handling Infinite Sequences:**
Generators can easily represent and produce infinite sequences of data, such as an endless stream of numbers, without encountering memory limitations because they don't attempt to store the entire sequence.

**Pipelining and Data Processing:**
Generators can be chained together to form efficient data processing pipelines. Each generator in the pipeline can perform a specific transformation or filtering step on the data, passing the results to the next generator incrementally, which is beneficial for complex data manipulation tasks.

**Simplified Iterative Logic:**
Generators simplify the implementation of iterators and sequence generation. The yield keyword provides a more concise and readable way to define iteration logic compared to manually managing state in a class-based iterator.

**Improved Performance (in certain cases):**
By yielding values one at a time and avoiding the creation of large intermediate data structures, generators can lead to performance improvements, especially when only a subset of the generated values is consumed.

Ans)**A lambda** function in Python is an anonymous, small, single-expression function defined using the lambda keyword. Unlike regular functions defined with def, lambda functions do not have a name and are restricted to a single expression, which is implicitly returned.

Higher-Order Functions: They are frequently used with higher-order functions like map, filter, and sorted, where a small function needs to be passed as an argument.

Q9)Explain the purpose and usage of the `map()` function in Python?

Ans)The map() function in Python is a built-in higher-order function that applies a given function to each item in an iterable (like a list, tuple, set, or string) and returns an iterator that yields the results.
Purpose:

**Conciseness and Readability:**
It provides a more compact and often more readable way to perform element-wise transformations on iterables compared to traditional for loops, especially for simple operations.

**Functional Programming:**
It aligns with functional programming paradigms by emphasizing the application of functions to data without modifying the original data structure directly.

**Efficiency (in Python 3):**
In Python 3, map() returns an iterator, which means it processes elements lazily, consuming less memory for large iterables compared to creating a new list immediately.

Example:

# Applying a function to a single list
    numbers = [1, 2, 3, 4, 5]
    squared_numbers = map(lambda x: x**2, numbers)
    print(list(squared_numbers)) # Output: [1, 4, 9, 16, 25]

# Applying a function to multiple lists
    list1 = [1, 2, 3]
    list2 = [10, 20, 30]
    sum_of_elements = map(lambda x, y: x + y, list1, list2)
    print(list(sum_of_elements)) # Output: [11, 22, 33]

# Using a built-in function
    strings = ["hello", "world"]
    upper_strings = map(str.upper, strings)
    print(list(upper_strings)) # Output: ['HELLO', 'WORLD']

Q10)What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
Ans)The map(), filter(), and reduce() functions in Python are built-in tools for functional programming, each serving a distinct purpose in data transformation and manipulation:

**map():** This function applies a given function to each item in an iterable (like a list or tuple) and returns an iterator that yields the results. It is used for transformation, creating a new sequence where each element is the result of applying the function to the corresponding element of the original sequence. The length of the output sequence from map() is always the same as the input sequence.

 Example of map()
    numbers = [1, 2, 3, 4]
    squared_numbers = list(map(lambda x: x * x, numbers))
    # squared_numbers will be [1, 4, 9, 16]

**filter():** This function constructs an iterator from elements of an iterable for which a given function returns True. It is used for selection or filtering, creating a new sequence containing only the elements that satisfy a specific condition. The length of the output sequence from filter() is typically less than or equal to the input sequence.

 Example of filter()
    numbers = [1, 2, 3, 4, 5, 6]
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    # even_numbers will be [2, 4, 6]

**reduce():** This function, found in the functools module, applies a function of two arguments cumulatively to the items of an iterable, from left to right, so as to reduce the iterable to a single value. It is used for aggregation or reduction, combining all elements into a single result.  

    from functools import reduce

    # Example of reduce()
    numbers = [1, 2, 3, 4]
    sum_of_numbers = reduce(lambda x, y: x + y, numbers)
    # sum_of_numbers will be 10

In summary, map() transforms elements, filter() selects elements, and reduce() aggregates elements into a single value.

Q1) 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.

```python
def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum
```

Q4)4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check up to square root of n
        if n % i == 0:
            return False
    return True
```

Q5)5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class Fibonacci:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self  # Iterator object returns itself

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration  # End the iteration

        if self.count == 0:
            self.count += 1
            return self.a
        elif self.count == 1:
            self.count += 1
            return self.b
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
```

Q6)6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(max_exponent):
    for i in range(max_exponent + 1):
        yield 2 ** i
```

Q7)7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_lines(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.rstrip('\n')
```

Q9)9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Use map with a lambda to convert to Fahrenheit
fahrenheit_temps = list(map(lambda c: (c * 9/5) + 32, celsius_temps))

# Output the result
print("Fahrenheit temperatures:", fahrenheit_temps)
```

Q10) Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, input_str))
```
